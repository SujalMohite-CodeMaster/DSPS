# Creating a normalized histogram of the data set using Matplotlib and Seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\mohit\OneDrive\Documents\DSPS\DSPS\car_price_dataset.csv"
```

```python
df = pd.read_csv(file_path)
```

```python

# Set figure size
plt.figure(figsize=(10, 6))

# Matplotlib Histogram (Raw Frequency)
plt.hist(df['Year'], bins=20, color='lightcoral', edgecolor='black', alpha=0.6, label="Matplotlib Histogram")

# Seaborn Histogram with Density Normalization
sns.histplot(df['Year'], bins=20, kde=True, stat="density", color='blue', alpha=0.5, label="Seaborn Histogram")

# Adding title and labels
plt.title('Distribution of Car Model Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
